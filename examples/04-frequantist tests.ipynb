# Permutation test

```python
%matplotlib inline
import pandas as pd

from abtools.frequentist import permutation_test
```

```python
data = pd.read_csv('data.csv')
```

```python
a = data.query('abgroup == "control"')['sp']
b = data.query('abgroup == "test"')['sp']
```

```python
p_value, null_rejected = permutation_test(a, b)
```
